# Random Forest

Due to overfitting in decision tree we use the random forest. Random forest is the model in which we make different different tree using bootstrap dataset & lastly check our predictor feature with each of the tree the majority will be our output

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/sharm/Desktop/PY/Random_forest/house-votes-84.csv",index_col=0)
df
```

```python
df = df.reset_index() #In our data frame we have ? which are null values
df
```

```python
df = df.replace("?",np.nan) #We will not be able to find null values by df.isna().sum() so we first need to convert them into nan
df
```

```python
df.columns
```

```python
republican = df[df["Class Name"]=="republican"]
republican
```

```python
republican= republican.replace("?",np.nan)
```

```python
republican.isna().sum()
```

```python
democrat = df[df["Class Name"]=="democrat"]
democrat
```

```python
democrat.isna().sum()
```

```python
for i in df.columns:
    republican[i] = republican[i].fillna(republican[i].mode()[0])
```

```python
for i in df.columns:
    democrat[i] = democrat[i].fillna(democrat[i].mode()[0])
```

```python
republican.isna().sum()
```

```python
democrat.isna().sum()
```

```python
new_df = pd.concat([republican,democrat],axis=0)
new_df
```

```python
y=new_df["Class Name"]
```

```python
x=new_df.drop("Class Name",axis=1)
x
```

```python
x.isna().sum()
```

```python
x.shape
```

```python
type(x)
```

```python
x.info()
```

```python
# our feature are not in numeric form so i need to convert them into numeric form with one hot or dummy encoding. so we know
# for non linear model we use one hot encoding & for linear models we use dummy encodinhg. random forest is non linear model so we will use one hot encoding
```

```python
x = pd.get_dummies(x,columns=x.columns)
x
```

```python
# splitting ourdataset into train test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42,stratify=y)
```

```python
# here all features are on same scale so we do not need feature scaling
```

```python
# so in theory we have noticed that these tree are based on boot strap dataset. so there are also out of bag data which was 
# not considered for our testing data set so if we want to check that we need to mention parameter as oob_score.
# we can also check our out of bag dataset core by .oob_score_
```

```python
from sklearn.ensemble import RandomForestClassifier  # random forest by default create 100 trees, if we want to custamize trees we use n_estimators parameter
forest = RandomForestClassifier(n_estimators=20,oob_score=True)
forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
forest.oob_score_  # here we check our out of bag dataset score which is 95%
```
